```python
import feast
from feast import FeatureStore
import yaml

with open('feature_store_feast_banking_client.yaml', 'r') as file:
    fs_config_yaml = yaml.safe_load(file)

fs_config = feast.repo_config.RepoConfig(**fs_config_yaml)
fs = feast.FeatureStore(config=fs_config)
```

```python
features = fs.get_online_features(
    entity_rows=[{"customer_id": "CUST_000001"}],
    features=[
        "transaction_7d_aggregations:amount",
        "transaction_7d_aggregations:transaction_type"
    ]
).to_dict()
print(features)
```
